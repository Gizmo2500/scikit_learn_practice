## 0. An end-to-end Scikit-Learn workflow

```python
# 1. Get the data ready
heart_disease = pd.read_csv("data/heart-disease.csv")

heart_disease
```

```python
#  Create X (features matrix)
X = heart_disease.drop("target", axis=1)

# Create y (labels)

y = heart_disease["target"]
```

```python
# 2.Choose the right model and hyperparameters (this case: classification)
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)

# We will keep the default hyperparameters 
clf.get_params()
```

```python
# 3. fit the model to the training data 
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clf.fit(X_train, y_train)
```

```python
# make a prediction (doesn't work)
# y_label = clf.predict(np.array([0, 2, 3, 4])) the array has to match data set size) 
```

```python
X_train
```

```python
# use the X_test for y_preds
y_preds = clf.predict(X_test)
y_preds
```

```python
y_test
```

```python
# 4. Evaluate the model on the training data and test data
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
# 5. Improve a model
# Try different amount of n_estimators

np.random.seed(42)

for i in range(5, 100, 5):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accuracy on test set: {clf.score(X_test, y_test) * 100:.2f}%")
    print("")
```

```python
# 6. Save a model and load it
import pickle

#Save the file ("wb" = write binary)
pickle.dump(clf, open("random_forest_model_1.pkl", "wb"))
```

```python
# load a model (note "rb" = read binary)

loaded_model = pickle.load(open("random_forest_model_1.pkl", "rb"))

loaded_model.score(X_test, y_test)
```

```python
# Turn the catagories into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["Make", "Colour", "Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                 one_hot,
                                 categorical_features)],
                                 remainder="passthrough")

# transformed_X = transformer.fit_transform(X)
# transformed_X
```

```python
transformed_X = transformer.fit_transform(X)
transformed_X
```
